3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
def create_a_circle(r):
    x = np.arange(-r,r,0.001)
    y=np.sqrt(r**2-x**2)
    plt.figure(figsize=(5,5))
    plt.grid(True)
    plt.xlim(-r-1,r+1)
    plt.ylim(-r-1,r+1)
    plt.plot(x,y)
    plt.plot(x,-y)
    plt.show()
```

```python
create_a_circle(2)
```

```python
def create_a_elipse(a,b):
    x = np.arange(-a,a,0.001)
    y=b/a*np.sqrt(a**2-x**2)
    plt.figure(figsize=(5,5))
    plt.grid(True)
    if a>b:      
        plt.xlim(-a-1,a+1)
        plt.ylim(-a-1,a+1)
    else:
        plt.xlim(-b-1,b+1)
        plt.ylim(-b-1,b+1) 
    plt.plot(x,y)
    plt.plot(x,-y)
    plt.show()
```

```python
create_a_elipse(5,4)
```

```python
def create_a_giperbola(a,b):
    x = np.arange(-a,a,0.001)
    y=b/a*np.sqrt(a**2+x**2)
    plt.figure(figsize=(5,5))
    plt.grid(True)
    if a>b:      
        plt.xlim(-a-1,a+1)
        plt.ylim(-a-1,a+1)
    else:
        plt.xlim(-b-1,b+1)
        plt.ylim(-b-1,b+1) 
    plt.plot(x,y)
    plt.plot(x,-y)
    plt.show()
```

```python
create_a_giperbola(2,1)
```

Решите систему уравнений:
y = x2 – 1

exp(x) + x∙(1 – y) = 1


Решите систему уравнений и неравенств:
    
y = x2 – 1

exp(x) + x∙(1 – y) > 1


```python
x=np.arange(-4,5,0.1)
y=x**2-1
y2=1-((1-np.exp(x))/x)
plt.grid(True)
plt.plot(x,y)
plt.plot(x,y2) 
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) +x*(1-y)-1)

x1, y1 =  fsolve(equations, (-2, 5))
x2, y2 =  fsolve(equations, (2, 5))
x3, y3 =  fsolve(equations, (4, 15))
print (f'x1= {x1}, y1= {y1}')
print (f'x2= {x2}, y2= {y2}')
print (f'x3= {x3}, y3= {y3}')
```
